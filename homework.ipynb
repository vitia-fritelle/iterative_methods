## Exercício 10

```python
def create_matrix_10_a(i: int, j: int, n: int):

    if i == j:
        if i == 0:
            return 12
        elif i == n-1:
            return -12
        else:
            return 6
    elif i == j-1 or i == j+1:
        if i == 0:
            return -6
        elif i == n-1:
            return 6
        else:
            return -4
    elif i == j-2 or i == j+2:
        if i == 0:
            return 4/3.0
        elif i == n-1:
            return 4/3.0
        else:
            return 1
    else:
        return 0
```

```python
def create_matrix_10_b(i: int, j: int, n: int):

    if i == j:
        if i == 0:
            return 12
        elif i == n-1:
            return 12/25.0
        elif i == n-2:
            return 111/25.0
        else:
            return 6
    elif i == j-1 or i == j+1:
        if i == 0:
            return -6
        elif i == n-1:
            return 24/25.0
        elif i == n-2:
            if j == i-1:
                return -93/25.0
            else:
                return -43/25.0
        else:
            return -4
    elif i == j-2 or i == j+2:
        if i == 0:
            return 4/3.0
        elif i == n-1:
            return 12/25.0
        else:
            return 1
    else:
        return 0
```

```python
def solve_exercise_10_a(n: int):
    
    print(n)
    A = np.array(
        [[create_matrix_10_a(i,j,n) for j in range(n)] for i in range(n)]
    )
    print(f"Os autovalores da matriz são {np.linalg.eigvals(A)}")
    x = np.linspace(-1,1,n,True)
    x = (x-1)*(x+1)
    b = A.dot(x)
    print(f"Eu encontei como solução o vetor: {conjugated_gradient_method(A.T.dot(A), A.T.dot(b))}")
    print(f"A solução encotrada pelo método np.linalg.solve {np.linalg.solve(A,b)}")
    print('')
    if n <= 100:
        solve_exercise_10_a(n+10)
```

```python
def solve_exercise_10_b(n: int):
    
    print(n)
    A = np.array(
        [[create_matrix_10_b(i,j,n) for j in range(n)] for i in range(n)]
    )
    print(f"Os autovalores da matriz são {np.linalg.eigvals(A)}")
    x = np.linspace(0,1,n,True)
    b = A.dot(x*x)
    print(f"Eu encontei como solução o vetor: {conjugated_gradient_method(A.T.dot(A), A.T.dot(b))}")
    print(f"A solução encotrada pelo método np.linalg.solve {np.linalg.solve(A,b)}")
    print('')
    if n <= 100:
        solve_exercise_10_b(n+10)
```

```python
solve_exercise_10_a(5)
```

```python
solve_exercise_10_b(5)
```

## Exercício 11

```python
def create_matrix_11(i: int, j: int, n: int):

    if i == j:
        return 3
    elif i+j == n-1:
        return 0.5
    elif j == i+1 or i == j+1:
        return -1
    else:
        return 0
```

```python
def solve_exercise_11(n: int):
    
    print(n)
    A = np.array(
        [[create_matrix_11(i,j,n) for j in range(n)] for i in range(n)]
    )
    x = np.ones(n)
    b = A.dot(x)
    print(b)
    print(f"Eu encontrei como solução o vetor: {conjugated_gradient_method(A, b)}")
    print(f"A solução encotrada pelo método np.linalg.solve {np.linalg.solve(A,b)}")
    print('')
    if n <= 100:
        solve_exercise_11(n+10)

```

```python
solve_exercise_11(5)
```
